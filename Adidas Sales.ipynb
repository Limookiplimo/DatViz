```python
# Import packages
import pandas as pd
import warnings
```

```python
# Ignore warnings
warnings.filterwarnings('ignore')
```

```python
# Import data
df=pd.read_csv('raw_data.csv')
```

```python
df.head()
```

```python
df.dtypes
```

To do:
- remove dollar sign
- remove percent sign
- invoice date to date type
- units sold to int
- price, sales and profit to float
- rename columns
- rearrange columns

```python
# Remove dollar & percent sign
def strip_chars(val):
    val = str(val)
    val = val\
            .replace("$","")\
            .replace("%","")
    return val

df["UnitPrice"] = df["UnitPrice"].apply(strip_chars)
df["TotalSales"] = df["TotalSales"].apply(strip_chars)
df["OperatingProfit"] = df["OperatingProfit"].apply(strip_chars)
df["OperatingMargin"] = df["OperatingMargin"].apply(strip_chars)
```

```python
# Convert invoice date
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Strip comma on units sold 
df['UnitsSold'] = df['UnitsSold'].str.replace(',', '').astype(int)
```

```python
# Convert to float
df['UnitPrice'] = df['UnitPrice'].astype(float)
df['TotalSales'] = df['TotalSales'].str.replace(',', '').astype(float)
df['OperatingProfit'] = df['OperatingProfit'].str.replace(',', '').astype(float)
```

```python
# Rename columns
df.rename(columns = {'UnitPrice':'Price($)', 'UnitsSold':'Units', 'TotalSales':'Sale($)','OperatingProfit':'Profit($)','OperatingMargin':'Margin(%)','SalesMethod':'Channel'}, inplace = True)
```

```python
#Add country column
df['Country'] = 'United States'
```

```python
# Re-arrange columns
new_cols= ['RetailerID', 'Retailer', 'Product', 'Country', 'Region', 'State', 'City', 'InvoiceDate', 'Price($)', 'Units', 'Sale($)', 'Profit($)', 'Margin(%)', 'Channel']
df = df.reindex(columns=new_cols)
```

```python
# Save date to csv
df.to_csv('clean_data.csv', index=False)
```
